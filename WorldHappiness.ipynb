```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
Here we can see 12% have outliers in data,as per study loosing a more data is not good for prediction,if data contains 5% to 8%
outliers we can remove the outliers 
So we cannot remove outliers in this data
```

```python
X=df.drop('Happiness Score',axis=1)
```

```python
This is Independent Variable
```

```python
X
```

```python
Y=df['Happiness Score']
```

```python
This is Label
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.30,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

ln=LinearRegression()
ln.fit(X_train,Y_train)
pred=ln.predict(X_test)
print(r2_score(Y_test,pred))
```

```python
The Logistic regression gives the 99% accuracy without cleaning data ,So sometimes loosing data is not good for predictions
Logistic Regression model is best fit for this dataset
```
